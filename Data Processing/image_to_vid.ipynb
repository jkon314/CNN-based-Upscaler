```python
##Convert Videos to still frame

import cv2


fPath = './Data/Frames/'

vid = cv2.VideoCapture("Data/CODGP_720.mp4")


```

```python
vidName = 'CODGP_720'
i = 1
while(True):

    ret,frame = vid.read()

    if ret:
        cv2.imwrite(f'{fPath}{vidName}_frame_{i}.jpg',frame)

        i+=1
        print(f'Processing Frame {i}')
    else:
        break


vid.release()
cv2.destroyAllWindows()
```

```python
#Remove files when necessary

import os
i = 1
while(True):

    if(os.path.isfile(f'{fPath}{vidName}_frame_{i}.jpg')):
        os.remove(f'{fPath}{vidName}_frame_{i}.jpg')
        print(f'Removing image {i}', end='\r')
        i+=1
    else:
        break
```
